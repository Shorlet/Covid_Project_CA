# Analyse des données Chine - Italie - France
## Création du DataFrame sur les décès

```python
# Données des décès pour la Chine, l'Italie et la France
# Sélection des données sur la France Métropolitaine
death_start = death.loc[(death['Country/Region'].isin(['China', 'Italy','France'])) & (death['Province/State'].isnull()), : ]

death_ch = death.loc[death['Country/Region'].isin(['China']), : ].groupby(by = 'Country/Region').sum()

death_start = death_start.append(death_ch, sort = False)

death_start
```

## Représentation des décès pour la Chine, l'Italie et la France

```python
# Choix des variables à représenter
x = x = np.array(death_fr.columns[4:])
y1 = death_start.iloc[0, 4:]
y2 = death_start.iloc[1, 4:]
y3 = death_start.iloc[2, 4:]

# multiple line plot
# Définition de la taille de la figure
plt.figure(figsize = (10,8))

# Dessin des lignes
plt.plot( x, y1, marker='', color='blue', linewidth=2, label = 'France')
plt.plot( x, y2, marker='', color='green', linewidth=2, label = 'Italie')
plt.plot( x, y3, marker='', color='red', linewidth=2,  label="Chine")

# Choix des titres
plt.title('Evolution des Décès du Covid19', fontweight = 'bold', fontsize = 20)
plt.xlabel('Temps', fontweight = 'bold', fontsize = 14)
plt.ylabel('Nombre de décès totaux', fontweight = 'bold', fontsize = 14)

# Création de la légende
plt.legend()

# Enregistrement du graphique
plt.savefig('Deces_debut_Cov19.jpg')

# Affichage du graphique
plt.plot()
```
